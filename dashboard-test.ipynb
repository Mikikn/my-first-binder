## Bar Chart

This example shows a bar chart with different colors for each bar.


```python
import matplotlib.pyplot as plt

# Make data
fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
bar_labels = ['red', 'blue', 'red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']

# Plot:
fig, ax = plt.subplots(figsize=(10, 6))  # Adjusted figure size

bars = []

for i in range(len(fruits)):
    bars.append(ax.bar(fruits[i], counts[i], label=bar_labels[i], color=bar_colors[i]))

ax.set_ylabel('Fruit supply')
ax.set_title('Fruit supply by kind and color')

# Adjust layout
ax.legend(title='Fruit color', bbox_to_anchor=(1.05, 1), loc='upper left')  # Moved legend outside
fig.tight_layout(pad=3.0)  # Adjusted padding

plt.show()
```

## Scatter Plot

This example shows a scatter plot with varying marker sizes and colors.


```python
import matplotlib.pyplot as plt
import numpy as np

# make data
np.random.seed(19680801)

# plot:
fig, ax = plt.subplots()
for color in ['tab:blue', 'tab:orange', 'tab:green']:
    n = 750
    x, y = np.random.rand(2, n)
    scale = 200.0 * np.random.rand(n)
    ax.scatter(x, y, c=color, s=scale, label=color, alpha=0.3, edgecolors='none')

ax.legend()
ax.grid(True)

plt.show()
```

## Histogram

This example shows a histogram of 1000 random values.



```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# make data
np.random.seed(1)
x = 4 + np.random.normal(0, 1.5, 200)

# plot:
fig, ax = plt.subplots()

ax.hist(x, bins=8, linewidth=0.5, edgecolor="white")

ax.set(xlim=(0, 8), xticks=np.arange(1, 8), ylim=(0, 56), yticks=np.linspace(0, 56, 9))

plt.show()
```
